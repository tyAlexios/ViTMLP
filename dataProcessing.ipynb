```python
import torch
import clip
from PIL import Image
import torch.nn.functional as F
import os
import random
import numpy as np
```

```python
datadir = '../dataset0/cam0'
Min = 10000
Max = -1
cnt = 0

ids = os.listdir(datadir)
random.shuffle(ids)

for i, id in enumerate(ids):
    views = os.listdir(os.path.join(datadir, id))
    len_views = len(views)
    if len_views >= 16:
        cnt += 1
    Min = min(len_views, Min)
    Max = max(len_views, Max)
    print(id, len_views)

print(Min, Max, cnt)
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
model, preprocess = clip.load("ViT-B/32", device=device)
```

```python
with torch.no_grad():
    chosen = []
    data = []
    
    ids = os.listdir(datadir)
    random.shuffle(ids)

    for i, ID in enumerate(ids):
        views = os.listdir(os.path.join(datadir, ID))
        random.shuffle(views)
        if len(views) < 16:
            continue
        chosen.append(int(ID))
        print('load id {}:'.format(ID), end = ' ')
        
        cnt = 0
        features = []
        for view in views:
            cnt += 1
            if cnt > 16:
                break
            print('  {0}: {1}'.format(cnt, view), end=' ')
            path = os.path.join(datadir, ID, view)
            image = preprocess(Image.open(path)).unsqueeze(0).to(device)
            f = model.encode_image(image)
            f = f.cpu().numpy()
            features.append(f)
        data.append(np.array(features))
        print("\n\n")

    print('DONE!')
    chosen = torch.Tensor(chosen)
    data = np.array(data)
    data = torch.from_numpy(data)
    data = data.squeeze()
    torch.save(chosen, 'chosenTensor.pkl')
    torch.save(data, 'dataTensor.pkl')
```
